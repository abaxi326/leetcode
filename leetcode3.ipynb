Given a binary array nums and an integer goal, return the number of non-empty subarrays with a sum goal.

A subarray is a contiguous part of the array.

```python
nums = [1,0,1,0,1]
goal = 2

i,j=0,0
sum=0
sol=0
while(j<len(nums)):
    sum+=nums[j]
    if sum==goal:
        sol+=1
    while(sum>=goal):
        sum-=nums[i]
        if sum==goal:
            sol+=1
        i+=1
    
    j+=1

sol
```
